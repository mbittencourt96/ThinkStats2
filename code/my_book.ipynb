### CAP 02

```python
import thinkstats2
```

```python
hist = thinkstats2.Hist([1,2,2,3,5])
```

```python
hist
```

```python
hist.Freq(2)
```

```python
hist[2]
```

```python
hist.Freq(4)
```

```python
hist.Values()
```

```python
for val in sorted(hist.Values()):
    print(val,hist.Freq(val))
```

```python
for val, freq in hist.Items():
    print(val,freq)
```

```python
import thinkplot
```

```python
thinkplot.Hist(hist)
thinkplot.Show(xlabel='Value',ylabel="Frequency")
```

```python
preg = nsfg.ReadFemPreg()
```

```python
live = preg[preg.outcome == 1]
```

```python
hist = thinkstats2.Hist(live.birthwgt_lb,label='birthwgt_lb')
```

```python
thinkplot.Hist(hist)
thinkplot.Show(xlabel="Pounds",ylabel="Frequency")
```

### Above: Gaussian or Normal Distribution

```python
hist_oz = thinkstats2.Hist(live.birthwgt_oz,label='birthwgt_oz')
thinkplot.Hist(hist_oz)
thinkplot.Show(xlabel="Ounces",ylabel="Frequency")
```

```python
hist_age = thinkstats2.Hist(live.agepreg,label='agepreg')
thinkplot.Hist(hist_age)
thinkplot.Show(xlabel="Age",ylabel="Frequency")
```

```python
hist_lgt = thinkstats2.Hist(live.prglngth,label='prglngth')
thinkplot.Hist(hist_lgt)
thinkplot.Show(xlabel="Weeks",ylabel="Frequency")
```

```python
firsts = live[live.birthord == 1]
others = live[live.birthord != 1]
```

```python
first_hist = thinkstats2.Hist(firsts.prglngth,label='first')
other_hist = thinkstats2.Hist(others.prglngth,label='other')
```

```python
width = 0.45
thinkplot.PrePlot(2)
thinkplot.Hist(first_hist,align='right',width=width)
thinkplot.Hist(other_hist,align='left',width=width)
thinkplot.Show(xlabel='weeks',ylabel='frequency',xlim=[27,46])
```
